### Your Code

```python
# The imports
import geopandas as gpd
import matplotlib.pyplot as plt

#loading in the GeoJson DataSet
gdf = gpd.read_file("data.geojson")

# Fixing the CRS so i can calculate the are GeoJSON files, since this file is in WGS84 which is lat/long
gdf = gdf.to_crs(epsg=32616)  # Converting the CRS to meters

# Calculating the surface area in square meters from the converted map (epsg:32616).
gdf['area_m2'] = gdf.geometry.area

# Restricting the yield values in the dictionary 
yield_mapping = {
    "Tennessee": 7.5, "Kentucky": 8.2, "Alabama": 6.8, "Georgia": 7.0

}
# Calculating the hectares (1 H is equal to 10k sq meters), needed to convert since the map coords were in meters and the given yield values are in hectares. "tons per hectare" so the math had to add up.
gdf['Land Area'] = gdf.geometry.area / 10000

# Shortcut for a manual for loop (How? : This line looks up each state name in the dictionary and copies the matching yield value into a new column for that row.)
gdf['yield_per_hectare'] = gdf['name'].map(yield_mapping)

# Filtering the data, by making a subset (' .isin ' is the/a filter)
gdf_filtered = gdf[gdf["name"].isin(yield_mapping.keys())].copy()

#Calculating total harvest by going row by row. 
gdf_filtered['total_harvest_tons'] = gdf_filtered['Land Area'] * gdf_filtered['yield_per_hectare']

# Results
print(gdf_filtered[['name', 'Land Area', 'total_harvest_tons']])





# To get a better visual I choose to make a bar graph so I can see a comparison up close and not miss details and to form good observations 

# To make it in order
gdf_filtered = gdf_filtered.sort_values('total_harvest_tons', ascending=False)
# Set the figure size
plt.figure(figsize=(6, 7))

# Create the bar chart from the filtered data
bars = plt.bar(gdf_filtered['name'], gdf_filtered['total_harvest_tons'], color='forestgreen')

# Add data labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    # f"{yval:,.0f}" formats the number with commas and zero decimals (e.g., 152,431)
    plt.text(bar.get_x() + bar.get_width()/2, yval + (yval * 0.01), f"{yval:,.0f}", 
             ha='center', va='bottom', fontweight='bold', fontsize=11)

# Add titles and labels
plt.title('Estimated Total Harvest by State', fontsize=14, pad=15)
plt.xlabel('States', fontsize=12)
plt.ylabel('Total Harvest (Tons)', fontsize=12)

# Added a light grid to make the values easier to track
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
